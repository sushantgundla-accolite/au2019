```python
# load the data
import pandas as pd
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# check names of feature columns
boston.feature_names
```

```python
# load the feature data
dfx = pd.DataFrame(boston.data, columns = boston.feature_names)
dfx.head(5)
```

```python
# load the target data
dfy = pd.DataFrame(boston.target, columns = ['target'])
dfy.head()
```

### Find Heat map of Correlation.(**Hint:** sns.heatmap(df))

```python
%matplotlib -l
```

```python
# import visualization packages
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline

# heatmap plot for correlation of the feature data
plt.figure(figsize = (12,10))
sns.heatmap(dfx.corr(),annot=True,cmap=sns.color_palette("coolwarm", 100))
```

####  Linear Regression for Target (Dont do train test split)

```python
# import Linear Regression model
from sklearn.linear_model import LinearRegression

# fit the model
model_lr = LinearRegression()
model_lr.fit(dfx,dfy)
```

```python
# print the intercept and the coeff values for the trained model
print(' Intercept:',model_lr.intercept_,'\n\n','Coeff list:',model_lr.coef_,'\n')
```

#### Test it on the Train set and find the RMSE of predicted and actual target.

```python
# import the error functions from sklearn
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

# predict the target values for entire train data
y_pred = model_lr.predict(dfx)
```

```python
# print the error values
from math import sqrt
print(' MAE on entire training set:',mean_absolute_error(y_pred,dfy))
print(' MSE on entire training set:',mean_squared_error(y_pred,dfy))
print(' RMSE on entire training set:',sqrt(mean_squared_error(y_pred,dfy)))
print(' R2 score on entire training set:',r2_score(y_pred,dfy))
```

#### Our model performs above average
